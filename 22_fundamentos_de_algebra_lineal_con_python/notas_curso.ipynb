### Documentaci칩n de dot():

Dot product of two arrays. Specifically,

- If both*a*마nd*b*마re 1-D arrays, it is inner product of vectors (without complex conjugation).
- If both*a*마nd*b*마re 2-D arrays, it is matrix multiplication, but using**[matmul](https://numpy.org/doc/stable/reference/generated/numpy.matmul.html#numpy.matmul)**맖r`a@막`말s preferred.
- If either*a*맖r*b*말s 0-D (scalar), it is equivalent to**[multiply](https://numpy.org/doc/stable/reference/generated/numpy.multiply.html#numpy.multiply)**마nd using`numpy.multiply(a,막)`맖r`a*막`말s preferred.
- If*a*말s an N-D array and*b*말s a 1-D array, it is a sum product over the last axis of*a*마nd*b*.
- **If*a*말s an N-D array and*b*말s an M-D array (where`M>=2`), it is a sum product over the last axis of*a*마nd the second-to-last axis of*b.***

## propiedades del producto punto: asociativa y distributiva, no es conmutativa

- asociativa: $$Ax(BxC)=(AxB)xC$$
- distribuitiva: $$AX(B+C) = (AxB) + (AxC)$$
- conmutativa: $$(AxB)  !=  (BxA)$$

No es conmutativa porque al invertir el orden de multiplicaci칩n las columnas de la una ya no ser치n iguales a las filas de la otra. Y aunque lo fueran, ser칤an operaciones totalmente distintas. Geom칠tricamente ser칤an transformaciones especiales distintas.

Por otro lado, los vectores, al tener solo una dimensi칩n, s칤 son conmutativos.

Demostraci칩n:

````python
A = np.arange(6).reshape(3,2)
B = np.arange(1,5).reshape(2,2)
C = np.arange(5,9).reshape(2,2)
````

````python
(A@B).T
````

````python
(B.T@A.T)
````

````python
# Comprobacion asociativa
A_BC = np.dot(A,B@C)
AB_C = np.dot(A@B,C)
print(A_BC == AB_C)
````

````python
# Comprobacion distributiva
D = np.dot(A,B+C)
E = np.dot(A,B) + np.dot(A,C)
print(D == E)
````

````python
# Comprobacion NO es conmutativa
F = np.dot(B,C)
G = np.dot(C,B)
print(F == G)
````

### ++ transformar una sistema de ecuaciones a matrices   
usaremos el sistema antes definido
$$y=3x+5 $$ 
$$y=2x+3$$
1. primero colocamos las variables del mismo lado de la ecuacion
$$-3x+y = 5 $$ 
$$-2x+y = 3 $$
2. ahora definimos cada coeficiente en una matriz 
\begin{equation}
\begin{bmatrix}
[-3,1]\\
[-2,1]
\end{bmatrix}
@
\begin{bmatrix}
[x,y]
\end{bmatrix}
= 
\begin{bmatrix}
[5,3]
\end{bmatrix}
\end{equation}  

date cuenta que el sistema de matrices tiene la forma 

$$[A]*[x]=[b]$$

    lo interesante aquie es quesi aplicamos el producto punto obtenemos las ecuaciones originales 

3. ahora solo queda despejar la matriz con las variables. donde $[A]^-1$ es la inversa
$$[x]=[b]*[A]^-1$$
 

````python
# definimos las matrices
A = np.array([[-3,1],[-2,1]])
b = np.array([[5],[3]])
sol_1 = np.array([-2,-1])
# vemos que al aplicar el producto punto obtenemos la solucion 
print(A.dot(sol_1))
````

### Espacio euclidiano o Espacio vectorial R1,R2,R3
Espacio euclidiano o Espacio vectorial es el conjunto de n-adas ordenadas, tambien conocido por espacio n-dimencional y de denota por Rn este es una sucesi칩n de n n칰meros reales ejemplo (a1,a2,,an) donde los vectores Rn se clasifican as칤:

- R3: R^3=RxRxR se refiere a un cubo 3D

## Espacio y un subespacio

- Espacio: donde est치 definido los vectores v1 y v2.
- Subespacio: donde est치 definido la combinaci칩n lineal de v1 y v2.
- Hiperplano: espacio de una dimensi칩n menor de donde est치n definidos v1 y v2.
    - en R3 un hiperplano es R2 o un plano carteciano
    - en R2 un hiperplano es R1 o un recta

siendo $$V= [V1,...,Vn]$$
- La norma L0: devuelve la cantidad de elementos diferentes de 0 del vector.  $$L0 : count(Vi) != 0$$
- La norma L1: devuelve la suma de los valores absolutos de los elementos del vector.  $$L1: sum(abs(Vi))$$
- La norma L2: La magnitud o distancia euclidiana del vector. distancia desde el origen hasta el vector. es calcular la hipotenusa de los vectores. es la mas usada. puede ser calculada de 2 maneras: 
    - usando trigonometria. es calcular la hipotenusa del teorema de pitagoras para n dimensiones: $$L2 = sqrt(v1^2+v2^2+...+vn^2)$$
    - calculando el producto interno del vector por el mismo: $$L2 = sqrt(V.T * V)$$
- La norma infinito: devuelve el valor absoluto m치s grande del vector

Para acceder a cada uno, solo se debe a침adir el argumento `ord=n`. Ejemplo: $$Li: max(abs(Vi))$$

~~~
vector = np.array([1,2,0,5,6,0])

np.linalg.norm(vector, ord=0) #norma 0 - devuelve el # de valores distintos de 0
np.linalg.norm(vector, ord=1) #norma 1 - suma de los valores absolutos
np.linalg.norm(vector, ord=2) #norma 2 (por defecto) - magnitud del vector
np.linalg.norm(vector, ord=2)**2 #norma 2^2
vector.T.dot(vector) #norma 2^2 (sale lo mismo que o de arriba)
print(np.linalg.norm(vector, ord=np.inf)) #norma infinito - max de los valores absolutos
~~~

Una matriz es ortogonal cuando todas sus filas y columnas son mutuamente ortonormales.

propiedad:
1. Una matriz ortogonal es una matriz cuadrada cuya matriz inversa coincide con su matriz transpuesta.
$$A.T@A = A@A.T = identidad$$
2. de lo anterior se deduce que:
$$A.T = A^-1$$


````python
matriz = np.diag([1,1,1]) # es lo mismo que np.eye(3, dtype=int)
matriz
````

````python
#comprobaremos que la matriz es ortogonal, 1ro comprobamos que todas las filas y columnas son ortogonales mutuamente 

# Si todas dan 0, es que todas sus columnas son ortogonales
print(matriz[:,0].dot(matriz[:,1]))
print(matriz[:,0].dot(matriz[:,2]))
print(matriz[:,1].dot(matriz[:,2]))

# Si todas dan 0, es que todas sus filas son ortogonales
print(matriz[0,:].dot(matriz[1,:]))
print(matriz[0,:].dot(matriz[2,:]))
print(matriz[1,:].dot(matriz[2,:]))
````

````python
# 2do comprobamos que todas las filas y columnas son ortonormales mutuamente 

# Si todas dan 1, es que todas sus columnas son ortonormales
print(np.linalg.norm(matriz[:,0]))
print(np.linalg.norm(matriz[:,1]))
print(np.linalg.norm(matriz[:,2]))

# Si todas dan 1, es que todas sus filas son ortonormales
print(np.linalg.norm(matriz[0,:]))
print(np.linalg.norm(matriz[1,:]))
print(np.linalg.norm(matriz[2,:]))
# concluimos que la matriz es ortogonal !!!!
````

## determinante 

el determinante es un numero que se asocia a una matriz que debe ser cuadrada. se obtiene al realizar una suma de los productos de cada elemento. si es una matriz diagonal se obtiene de multiplicar todos los elementos de la diagonal de la matriz. Nos permite simplificar operaciones matriciales como el calculo del rango o de la matriz inversa. otro de sus usos mas importantes es para saber si un sistema de ecuaciones tiene solucion, o dicho de otra forma, para saber si los vectores de la matriz A son independientes ya que supongamos tenemos el siguiente sistema de ecuaciones y lo queremos resolver: 
$$ax+by=c$$
$$dx+ey=f$$
1. hacemos que cada ecuacion solo contenga una incognita, haciendo despeje:  

$ax+b( \frac{(f-dx)}{e})=c -> ax+\frac{(bf-bdx)}{e}=c -> \frac{(bf-bdx+aex)}{e}=c -> -bdx+aex=ac-bf -> x(ae-bd)= ac-bf $    
resultando
$$x= \frac{ac-bf}{ae-bd}$$
haciendo lo mismo con la otra ecuacion obtenemos
$$x= \frac{af-dc}{ae-bd}$$
del sistema de ecuaciones 
$$ax+by=c$$
$$dx+ey=f$$
obtenemos la matriz A
\begin{equation}
\begin{bmatrix}
[a,b]\\
[d,e]
\end{bmatrix}
\end{equation}
